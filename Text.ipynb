# PROMPT TEMPLATE TEXT SUMMARIZATION-2ND WAY


```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
generic_template = "Write a summary of the following speech: {speech}\nTranslate the precise summary to {language}."
prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=generic_template
)
prompt
```

# STUFF DOCUMENT CHAIN-SUMMARIZATION

Stuff Document Chain Summarization
Stuff document chain summarization is the most basic summarization technique. Consider an external data source such as a PDF, text file, or website. The entire content of this source is read and combined into a single document.

-Challenges with Stuff Document Chain
While this method works well for smaller documents, it faces significant challenges with larger files. For instance, if a PDF contains 100 pages or more, combining all the text into one prompt becomes impractical.

```python
from langchain_community.document_loaders import PyPDFLoader
loader= PyPDFLoader("apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
template = '''
Write a concise and short summary of the following speech:
{text}
'''
from langchain.prompts import PromptTemplate
prompt = PromptTemplate(template=template, input_variables=['text'])
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain=load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
output_summary=chain.run(docs)
output_summary
```

# MAP REDUCE
- MapReduce Summarization Technique
To address the limitations of the stuff document chain, we use the MapReduce summarization technique, which is especially useful for larger documents.

How MapReduce Works--
- Divide the Document into Chunks: The large document is split into smaller, manageable chunks.
- Map Step: Each chunk is passed individually to a prompt template and then to the LLM to generate a summary for that chunk.
- Reduce Step: All the individual summaries are combined and passed through another prompt template to generate the final summary.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100)
final_documents = text_splitter.split_documents(docs)
```

```python
chunk_prompt = '''
Please summarize the below speech:
{text}
'''
map_prompt = PromptTemplate(template=chunk_prompt, input_variables=['text'])
```

```python
final_prompt = '''
Provide the final summary of the entire speech with these important points:
- Add a motivational title.
- Start the precise summary with an introduction.
- Provide the summary in numbered points for the speech.
{text}
'''
combine_prompt = PromptTemplate(template=final_prompt, input_variables=['text'])
```

```python
summary_chain = load_summarize_chain(
    llm,
    chain_type='map_reduce',
    map_prompt=map_prompt,
    combine_prompt=combine_prompt,
    verbose=True
)
output = summary_chain.run(final_documents)
```

# Refine Chain Summarization Process--
-- In refine chain summarization, we take the first chunk and pass it to a prompt template along with the language model to obtain a summary. Once we get this summary, it serves as a reference for the next chunk. Before sending the second chunk to the prompt and language model, we include the previous summary as context.

-- This process continues sequentially, where each chunk's summary is combined with the rolling summary and passed to the prompt and language model again to get a refined summary.

-- Refine summarization means updating a rolling summary by iterating over the document in sequence.

-- For every chunk, when passing it to the prompt, the existing summary is combined with the chunk. The language model then produces a new summary that refines the previous one. This rolling summary is updated chunk by chunk until all chunks are processed, resulting in the final summary.

```python
chain = load_summarize_chain(llm, chain_type="refine", verbose=True)
output_summary = chain.run(docs)
print(output_summary)
```
